# Introdução

## Objetivos

- Aprender como modelos matemáticos podem ser formulados com base em princípios científicos para simular o comportamento de um sistema físico simples.
- Entender como métodos numéricos oferecem um meio de generalizar soluções de uma forma que possa ser implementado em um computador digital.
- Conhecer os diferentes tipos de métodos numéricos que serão abordados nesta disciplina.

## Um modelo matemático simples

- Um modelo matemático pode ser amplamente definido como uma formulação ou equação que expressa as características essenciais de um sistema físico ou processo em termos matemáticos.
- Os modelos podem ser representados por uma relação funcional entre variáveis dependentes, variáveis independentes, parâmetros e funções de força.

## Modelagem de função

![](img/Picture1.png)

- Variavel dependente - uma característica que geralmente reflete o comportamento ou estado do sistema
- Variavel Independente - dimensões, tais como tempo e espaço, ao longo do qual o comportamento do sistema está sendo determinado
- Parâmetros - constantes que refletem propriedades e composição do sistema
- Forças externas - influências externas que agem sobre o sistema
